```python
import graphviz
import pandas as pd
from sklearn import tree
```

```python
df_GMM = pd.read_csv('data/clustered_data.csv', encoding='cp949')
```

```python
del df_GMM['영업이익률']
```

```python
X = df_GMM.iloc[:, :-1].to_numpy()
```

```python
X.shape
```

```python
Y = df_GMM['clusters']
```

```python
class_names = ['0', '1', '2']
```

```python
clf = tree.DecisionTreeClassifier(max_depth=20)
clf = clf.fit(X, Y)  
```

```python
df_feature = df_GMM
# del df_feature['clusters']
```

```python
# 시각화
dot_data = tree.export_graphviz(clf,   # 의사결정나무 모형 대입
                               out_file = None,  # file로 변환할 것인가
                               feature_names = df_feature.columns,  # feature 이름
                               class_names = class_names,  # target 이름
                               filled = True,           # 그림에 색상을 넣을것인가
                               rounded = True,          # 반올림을 진행할 것인가
                               special_characters = True)   # 특수문자를 사용하나

graph = graphviz.Source(dot_data)              
graph
```
