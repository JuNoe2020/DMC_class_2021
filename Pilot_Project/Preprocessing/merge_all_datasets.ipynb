```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/preprocessed_df_comp_demo.csv', encoding='cp949')
df2 = pd.read_csv('data/jobplanet_rates_data.csv', encoding='cp949')
df3 = pd.read_csv('data/labelled_jobplanet_review_for_RNN.csv', encoding='cp949')
```

```python
merged_df1 = pd.merge(df1, df2, how='inner', on='기업명')
```

```python
merged_df2 = pd.merge(merged_df1, df3, how='inner', on='기업명')
```
